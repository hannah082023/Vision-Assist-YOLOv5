## Data Splitting
Split dataset into 80% training and 20% validation

```python
import os
import math

def split_dataset(src_dir='dataset', split_rate=0.8):
    images_dir = os.path.join(src_dir, 'images2')
    if os.path.exists(images_dir):
        total_images = os.listdir(images_dir)
        train_count = math.floor(len(total_images) * split_rate)
        print(f"Total images: {len(total_images)}")
        print(f"Training images: {train_count}")
        print(f"Validation images: {len(total_images) - train_count}")
    else:
        print("Dataset directory not found. Please upload your dataset first.")

split_dataset()
```

## Generate Configuration YAML
Write tra.yaml configuration file for YOLOv5

```python
yaml_content = """
path: ../dataset          # Dataset root path
train: images/train       # Train images
val: images/valid         # Validation images

nc: 3                     # Number of classes
names: ['redlight', 'greenlight', 'zebralines']
"""

with open('tra.yaml', 'w') as f:
    f.write(yaml_content.strip())
print("tra.yaml configuration file has been written successfully.")
```
